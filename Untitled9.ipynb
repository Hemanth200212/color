```python
import cv2
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from google.colab import files
```

```python
uploaded = files.upload()

# Load the first uploaded file
image_path = next(iter(uploaded))
print(f"Loaded image: {image_path}")
```

```python
def load_image(image_path):
    image = cv2.imread(image_path)
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

image = load_image(image_path)
plt.imshow(image)
plt.title("Original Image")
plt.axis("off")
plt.show()

```

```python
def get_dominant_colors(image, k=5):
    img = cv2.resize(image, (100, 100))  # Resize for faster processing
    img = img.reshape((img.shape[0] * img.shape[1], 3))
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(img)
    return kmeans.cluster_centers_, kmeans.labels_

colors, labels = get_dominant_colors(image, k=5)
```

```python
def plot_dominant_colors(colors, labels):
    unique_labels, counts = np.unique(labels, return_counts=True)
    colors = colors.astype(int)

    hex_colors = ['#{:02x}{:02x}{:02x}'.format(r, g, b) for r, g, b in colors]

    plt.figure(figsize=(8, 4))
    plt.pie(counts, labels=hex_colors, colors=np.array(colors)/255, startangle=90)
    plt.title("Dominant Colors")
    plt.axis("equal")
    plt.show()

plot_dominant_colors(colors, labels)
```
